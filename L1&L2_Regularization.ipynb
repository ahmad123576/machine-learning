```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# suppress WARNINGS for clean notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/Melbourne_housing_FULL.csv')
df.head()
```

```python
df.nunique()
```

```python
# lets use limited columns
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=df[cols_to_use]
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
# handling missing values
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[cols_to_fill_zero]=df[cols_to_fill_zero].fillna(0)

# features can be imputed with mean for faster results
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.shape
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
# we don't impute price b/c its our predictive variable
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
# lets do linear regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

```python
# here training score is 68% and test score is 13% which is very low, So:
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
# its good, now check with ridge:
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
